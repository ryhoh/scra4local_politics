```python
import pickle
import pandas as pd
```

```python
with open('../res.pkl', 'rb') as f:
    dicts = pickle.load(f)
dicts
```

```python
import sys
sys.path.append('../')
from src.crawler import frame_unpack
from src.parser import parse
```

```python
dicts2 = dicts.copy()
for d in dicts2:
    d['attend'] = parse(frame_unpack(d['HTML']))
dicts2
```

```python
attendance = pd.DataFrame({'council_name': [], 'number': [], 'person_name': []})
attendance
```

```python
for d in dicts2:
    for person in d['attend']:
        attendance = attendance.append({
            'council_name': d['会の名称'],
            'number': person['number'],
            'person_name': person['name']
        }, ignore_index=True)

attendance
```

```python
attendance.to_csv('高槻市_20201104収集.csv')
```
